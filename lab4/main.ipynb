**Далі видаляємо рядкі що містять в собі "Unknown" значення відносно колонок "Genres", "Type", "Rating", "Duration"**

```python
df = df.drop(df.loc[ ((df["Genres"]=="Unknown")|(df["Type"]=="Unknown")) | ((df["Rating"]=="Unknown")|(df["Duration"]=="Unknown") ) ].index, axis=0)
df.isin(["Unknown"]).sum()
```

**Заповнимо "Unknown" значення колонки "Ranked" та "Score" задопомогою лінійної інтерполяції**
**Формула лінійної інтерполяції:**
$$ y = y_{1} + \frac{ y_{2} - y_{1} }{ x_{2} - x_{1} }\cdot (x - x_{1}) $$

```python
df.loc[df.loc[df["Ranked"]=="Unknown"].index, "Ranked" ] = np.nan
df["Ranked"] = df["Ranked"].astype(float).interpolate().astype(int)

df.loc[df.loc[df["Score"]=="Unknown"].index, "Score" ] = np.nan
df["Score"] = df["Score"].astype(float).interpolate()
```

```python
df.isin(["Unknown"]).sum()
```

<h1 align="center">Додаткове завдання</h1>

#### One hot encoding стовпця "Genres"

```python
genres = df["Genres"].copy().to_frame()
genres_list = sorted( list(set( df["Genres"].apply(lambda x: x+", ").sum().split(sep=", ") ))[1:] )


for genre in genres_list:
    genre_time_series = df["Genres"].apply(lambda x: 1 if genre in x else 0).to_frame(genre)
    genres = pd.concat([genres, genre_time_series], axis=1)

genres.head(3)
```

#### Спроба на OHE стовпця "Studios"

```python
sudios = df["Studios"].copy()
studios_list = sorted( list(set( df["Studios"].apply(lambda x: x+", ").sum().split(sep=", ") ))[1:] )

len(studios_list)
```

_Мій старенький ноутбук не готовий до такого, тому було приянято рішення видалити цю колонку..._

#### Підготовка data frame до розбиття

```python
df_y = df["Score"].copy()
df_x_without_genres  = df.drop(["Studios", "Name", "Score", "Genres"], axis=1)
df_x = pd.concat([df_x_without_genres, genres.iloc[:, 1:]], axis=1)
```

```python
display( np.unique(df_x["Type"]).shape )
display( np.unique(df_x["Source"]).shape )
display( np.unique(df_x["Duration"]).shape)
display( np.unique(df_x["Rating"]).shape)
```

#### Ordinal encoding стовпців "Type", "Source", "Duration", "Rating"

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
df_x[["Type", "Source", "Duration", "Rating"]] = oe.fit_transform(df_x[["Type", "Source",
                                                                        "Duration", "Rating"]])
df_x.head()
```

#### Розбиття data frame на дві рівні частини

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, train_size=0.5, test_size=0.5)
```

#### Лінійна регресія

```python
from sklearn import linear_model

line_reg = linear_model.LinearRegression()

line_reg.fit(x_train, y_train)
line_reg_y_pred = line_reg.predict(x_test)
```

#### Хребет(Ridge)

```python
ridge = linear_model.Ridge(alpha=500.5)

ridge.fit(x_train, y_train)
ridge_y_pred = ridge.predict(x_test)
```

#### Elastic-Net

```python
elnet = linear_model.ElasticNet(alpha=5000.1)

elnet.fit(x_train, y_train)
elnet_y_pred = elnet.predict(x_test)
```

#### Візуалізації моделей

```python
from yellowbrick.regressor import PredictionError, ResidualsPlot

fig, axes = plt.subplots(1, 3, figsize=(20, 10))

visualizer = [
    PredictionError(line_reg, ax=axes[0]),
    PredictionError(ridge, ax=axes[1]),
    PredictionError(elnet, ax=axes[2])
]

for viz in visualizer:
    viz.fit(x_train, y_train)
    viz.score(x_test, y_test)
    viz.finalize()

plt.show()
```

#### Середня квадратична похибка

```python
from sklearn.metrics import mean_squared_error

for i in [line_reg_y_pred, ridge_y_pred, elnet_y_pred]:
    display(mean_squared_error(y_test, i))
```

**Висновок : лінійна регресія найкраща**
