### Завдання №1

```python
# На 1000 000 сильно виснe, 1000 найприємніше значення :3
n = 1000
x = np.linspace(0, 1, n)
k = 50
b = 50
y_org = k*x+b
y_input = y_org + np.random.randint(-100, 100, n) * np.random.rand(n)*3*n/1001
```

```python
def plot_dot(lw_=3):
    plt.figure(figsize=(10,6))
    plt.scatter(x, y_input, label="Input value", color='black')
    plt.plot(x, y_org, label = "Original line", color="red", lw=lw_) 

plot_dot()
plt.legend(loc="best") 
```

```python
def least_square_method(x: np.array, y: np.array) -> (float, float):
    x_ = np.mean(x)
    y_ = np.mean(y)
    b1 = np.sum((x-x_)*(y - y_)) / np.sum(np.power((x - x_), 2))
    b0 = y_ - b1*x_
    return [b1, b0]
```

```python
print(f'My func:{least_square_method(x, y_input)}\nNumpy func:{np.polyfit(x, y_input, 1)}')
```

```python
my_b1, my_b0 = least_square_method(x, y_input)
np_b1, np_b0 = np.polyfit(x, y_input, 1)
plot_dot(10)
lw_ = 10
plt.plot(x, my_b1 * x + my_b0, label="My lsm", lw=lw_, color='blue')
plt.plot(x, np_b1 * x + np_b0, label="Numpy", lw=lw_, ls='--', color='green')
plt.legend()
```

### Завдання №2

```python
def mse(x: np.array, y: np.array, b0: float, b1: float) -> float:
    return np.sum(np.power(y - b0 - b1 * x, 2))/n
```

```python
mse(x, y_input, least_square_method(x, y_input)[1], least_square_method(x, y_input)[0])
```

```python
def gradient(x: np.array, y: np.array, learning_rate=.6, n_iter=100):
    b0 = 10
    b1 = 10
    m = len(x)
    for i in range(n_iter):
        y_ = b0 + b1 * x
        
        dL_db0 = -2 * np.sum(y - y_) / n
        dL_db1 = -2 * np.dot(x, (y - y_)) / n

        b0 -= learning_rate * dL_db0
        b1 -= learning_rate * dL_db1

    # print(f'MSE:{mse(x, y_input, b0, b1)}')

    return [b0, b1]
```

```python
def gradient_costs(x: np.array, y: np.array, learning_rate=.1, n_iter=100):
    b0 = 10
    b1 = 10
    costs = np.empty(n_iter)
    for i in range(n_iter):
        y_ = b0 + b1 * x
        
        dL_db0 = -2 * np.sum(y - y_) / n
        dL_db1 = -2 * np.dot(x, (y - y_)) / n

        b0 -= learning_rate * dL_db0
        b1 -= learning_rate * dL_db1

        costs[i] = mse(x, y, b0, b1)

    return costs


def learning_rate_costs(x, y, learning_rate):
    costs = np.empty(len(learning_rate))
    for i in range(len(learning_rate)):
        b0, b1 = gradient(x, y, learning_rate[i])
        costs[i] = mse(x, y, b0, b1)

    return costs


find_iter = 1000
n_rate = 200
find_learning_rate = np.linspace(0, .1, n_rate)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13,6), sharey=True)
ax1.plot(np.arange(find_iter), gradient_costs(x, y_input, n_iter=find_iter) )
ax2.plot(find_learning_rate, learning_rate_costs(x, y_input, find_learning_rate) )
# Orange line
ax1.plot(np.arange(find_iter), [10500]*find_iter)
ax2.plot(find_learning_rate, [10500]*n_rate)
```

```python
plot_dot(10)
lw_ = 10

grad_b0, grad_b1 = gradient(x, y_input)
lsm_b1, lsm_b0 = least_square_method(x, y_input)

plt.plot(x, lsm_b1 * x + lsm_b0, label="Lsm", lw=lw_, color='blue')
plt.plot(x, grad_b0 + grad_b1 * x, label="Grad", lw=lw_, color="pink", ls='--')
```
